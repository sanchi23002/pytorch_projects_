```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace = True)
df.head()
```

```python
df.shape
```

**Test Train Split**



```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0], test_size=0.2)
```

**Scaling of the features(X)**

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
y_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

**Numpy  array to pytorch tensor**

```python
x_train = torch.from_numpy(x_train.astype(np.float32)) ## pytorch prefer float32 forcumputation of neural network
x_test = torch.from_numpy(x_test.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
```

```python
print(x_train.shape)
print(y_train.shape)
```

**Defining the model**

```python
import torch.nn as nn

class NeuralNetwork(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)## it is a fully connected neuron , take all features as input and 1 output
    self.sigmoid = nn.Sigmoid()

  def forward(self,features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

**Important parameters**

```python
learning_rate = 0.1
epoch = 100
```

**Loss function**

```python
loss = nn.BCELoss()
```

**Training pipeline**

```python
## always first create object(model)
model = NeuralNetwork(x_train.shape[1])

## defining optimizer
optimizer = torch.optim.SGD(model.parameters(),lr = learning_rate) ## batch gradient decent, w = w - n(x(i)*(ycap - y)) , b = b - n(ycap-y)

for i in range(epoch):
  y_pred = model(x_train)
  loss_found = loss(y_pred,y_train.view(-1,1))

  ##  gradient clearance
  optimizer.zero_grad()

  ## Backward Pass
  loss_found.backward()

  ##Parameter update
  optimizer.step()

  ## Print loss in each epoch.
  print("epoch",(i+1),"   loss: ",loss_found.item())
```

Evaluation

```python
## Model Evaluation
with torch.no_grad():
  y_pred = model(x_test)
  y_pred = (y_pred>=0.5).float()
  accuracy = (y_pred==y_test).float().mean()
  print("accuracy" , accuracy.item())
```
