```python
import torch
```

PyTorch has a facility where you just need to do forward propagation and it gives you the back propagation. Remember we can do back propagation only for the leaves. When you start the propagation you can't do the intermediate terms derivation by back propagation. You can only work with the first input.

```python
x = torch.tensor(3.0, requires_grad=True)
y = x**2
print(x,y)
```

```python
y.backward(retain_graph = True)
```

```python
dy_dx = x.grad
print(dy_dx)
```

**ONE NEURONE TRAINING**

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad = True)
```

```python
def binary_cross_entropy_loss(prediction, target):
  prediction = torch.clamp(prediction, min = 1e-9, max = 1-1e-9)
  pos_loss = -target * torch.log(prediction)
  neg_loss = -(1-target) * torch.log(1-prediction)
  total_loss = pos_loss + neg_loss
  return total_loss.mean()
```

```python
## forward pass
z = torch.mul(x,w) + b
prediction = torch.sigmoid(z)
print(prediction)
```

```python
loss = binary_cross_entropy_loss(prediction, y)
print(loss)
```

```python
loss.backward()
```

```python
dl_dw = w.grad
dl_db = b.grad
print(dl_dw, dl_db)
```

```python
x = torch.tensor([1.0,2.0,3.0], requires_grad=True)
y = (x**2).mean()
y.backward()
dy_dx = x.grad
print(dy_dx)
```
