```python
import numpy as np
import torch
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler## normalize the value in between 0 to 1
from sklearn.preprocessing import LabelEncoder## used for making binary level on the output
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head() ## show first five rows of the column
```

```python
## train test split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 1:],df.iloc[:, 0], test_size= 0.2)## iloc tell the location
## in both cases all rows are including for the study but for x we are taking from the second column , and for y we are taking only the first column
```

```python
## label_encoding
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)## fit tells what to lebel
y_test = encoder.transform(y_test)## transform tell follow the label that are taken by fit
##Number assign occurs alphabetically so B->benign is equal to 0 and M-> malignant is equal to one.
```

```python
y_train
```

```python
## scaling
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
## numpy array to pytorch tensor
x_train_tensor = torch.from_numpy(x_train)
x_test_tensor = torch.from_numpy(x_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
print(x_train_tensor.shape)
print(y_train_tensor.shape)
```

```python
## defining the model

class my_neural_network():
  def __init__(self,x):
    self.weights = torch.rand(x.shape[1],1,dtype = torch.float64, requires_grad= True)
  ## matrix multiplication is done in X and W , so W row must be equal to columns of x and the output of eash row is one prediction as it is binary classification
    self.bias = torch.ones(1,dtype=torch.float32, requires_grad=True)

  def forward(self,x):
    z = torch.matmul(x, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(slef, y_pred, y):
    epsilon = 1e-8
    y_pred = torch.clamp(y_pred, epsilon , 1 - epsilon)
    loss = -y*torch.log(y_pred) - (1-y)*torch.log(1-y_pred)
    mean_loss = torch.mean(loss)
    return mean_loss
```

```python
learning_rate = 0.01
epoch = 3000
```

```python
## training pipeline

# create model
model = my_neural_network(x_train_tensor)

for i in range(epoch):
  y_pred = model.forward(x_train_tensor)
  loss = model.loss_function(y_pred, y_train_tensor)
  loss.backward()
  with torch.no_grad():
    model.weights -= learning_rate*model.weights.grad
    model.bias -= learning_rate*model.bias.grad
  model.weights.grad.zero_()
  model.bias.grad.zero_()
  print(f'Epoch : {i+1}, loss : {loss.item()}')
```

```python
## Evaluation

with torch.no_grad():
  y_pred = model.forward(x_test_tensor)
  y_pred = (y_pred>0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(accuracy.item())
```
